```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/raw/Titanic-Dataset.csv')
df.shape
```

```python
df.info()
```

```python
df['Embarked'].value_counts()
```

```python
df['Survived_renamed'] = df['Survived'].apply(lambda x: 'Выжили' if x == 1 else 'Погибли')
```

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='Survived_renamed', data=df)
plt.title(f'Распределение пассажиров Титаника')
plt.xlabel('Погибли/Выжили')
plt.ylabel('Количество')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='Sex', hue='Survived_renamed', data=df)
legend = ax.get_legend()
if legend is not None:
    legend.set_title(None)
plt.title(f'Распределение выживших/погибших в зависимости от пола')
plt.xlabel('Пол')
plt.ylabel('Количество')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='Pclass', hue='Survived_renamed', data=df)
legend = ax.get_legend()
if legend is not None:
    legend.set_title(None)
plt.title(f'Распределение выживших/погибших в зависимости от класса')
plt.xlabel('Номер класса')
plt.ylabel('Количество')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='Embarked', hue='Survived_renamed', data=df)
legend = ax.get_legend()
if legend is not None:
    legend.set_title(None)
plt.title(f'Распределение выживших/погибших в зависимости от порта')
plt.xlabel('Порт')
plt.ylabel('Количество')
plt.show()
```

```python
df['Cabin_known'] = df['Cabin'].apply(lambda x: 'Нет' if type(x) == float else 'Да')
```

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='Cabin_known', hue='Survived_renamed', data=df)
legend = ax.get_legend()
if legend is not None:
    legend.set_title(None)
plt.title(f'Распределение выживших/погибших\nв зависимости от информации о каюте')
plt.xlabel('Есть ли информация о каюте?')
plt.ylabel('Количество')
plt.show()
```

```python
survived = df[df['Survived_renamed']=='Выжили']
died = df[df['Survived_renamed']=='Погибли']
```

```python
plt.figure(figsize=(6, 4))
ax = plt.hist(x='Age', data=survived[survived['Sex']=='male'], bins=30)
plt.title(f'Распределение выживших мужчин в зависимости от возраста')
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.ylim(0, 40)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
ax = plt.hist(x='Age', data=survived[survived['Sex']=='female'], bins=30)
plt.title(f'Распределение выживших женщин в зависимости от возраста')
plt.ylim(0, 40)
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
ax = plt.hist(x='Age', data=died[died['Sex']=='male'], bins=30)
plt.title(f'Распределение погибших мужчин в зависимости от возраста')
plt.ylim(0, 45)
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
ax = plt.hist(x='Age', data=died[died['Sex']=='female'], bins=30)
plt.title(f'Распределение погибших женщин в зависимости от возраста')
plt.ylim(0, 40)
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
ax = plt.hist(x='Age', data=died[died['Age']<18], bins=30)
plt.title(f'Распределение погибших детей в зависимости от возраста')
plt.ylim(0, 15)
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
ax = plt.hist(x='Age', data=survived[survived['Age']<18], bins=30)
plt.title(f'Распределение выживших детей в зависимости от возраста')
plt.ylim(0, 15)
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.show()
```
